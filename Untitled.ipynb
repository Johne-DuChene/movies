```python
# import cell
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sl
import tensorflow as tf
import xgboost as xgb
```

```python
# load dataframe and print head
df = pd.read_csv("movies.csv")
df.head()
```

```python
# basic info
df.info()
```

```python
# convert percentage column to float
df["RatingTomatometer"] = df["RatingTomatometer"].str.strip("%").astype("float") / 100
```

```python
df.head()
```

```python
# ohe genre
dummies = pd.get_dummies(df["Genre"])
dummies
```

```python
df = df.join(dummies)
```

```python
# drop text columns
df = df.drop(["Genre", "Title"], axis = 1)
```

```python
df.head()
```

```python
# plot correlation
sns.heatmap(df.corr())
```

```python
df.corr()
```

```python
# our target is the rating
target = "RatingTomatometer"
X = df.drop(target, axis = 1)
y = df[target]
X_train, X_test, y_train, y_test = sl.model_selection.train_test_split(X, y)
```

```python
# start with the xgboost model
model = xgb.XGBRegressor().fit(X_train, y_train)
model.predict(X_test)
X_test_results = pd.DataFrame(model.predict(X_test))
X_test_results
```

```python
X_test_results["actual"] = y_test
```
